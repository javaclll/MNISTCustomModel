```python
def convolutionLayer(inputData, kernel, imageSize):

    xKernel = kernel.shape[0]
    yKernel = kernel.shape[1]

    #finding the shape of the output Convoluted Result
    xoutputConvo = imageSize - xKernel + 1
    youtputConvo = imageSize - yKernel + 1

    outputConvo = np.zeros((xoutputConvo, youtputConvo))

    splittedInput = np.split(inputData, imageSize)
    splittedInput = np.asarray(splittedInput)
    for i in range(xoutputConvo):
        for j in range(youtputConvo):
            outputConvo[i,j] = (splittedInput[i:i+xKernel, j:j+yKernel] * kernel).sum()
            
    return outputConvo
```

```python
def poolingLayer(inputData, poolSize = 2):
    
    xinputData = inputData.shape[0]
    yinputData = inputData.shape[1]
    
    xoutputPool = math.floor(xinputData/poolSize)
    youtputPool = math.floor(yinputData/poolSize)
    
    outputPool = np.zeros((xoutputPool, youtputPool))
    
    #number of values used for average pooling
    noValPool = poolSize * poolSize
    
    for i in range(xoutputPool):
        for j in range(youtputPool):
            outputPool[i,j] = (inputData[i * poolSize:i * poolSize+poolSize, j * poolSize:j* poolSize+poolSize]).sum()/noValPool
    
    return outputPool
```

```python
def initialization(layers, X, Y):
    
    #nx is the input layer and ny is the output layer
    eachLayers = {
        "n_x" : X.shape[0],
        "n_y" : Y.shape[0]
    }
  
    i = 1
    
    # Layers could be [4,4] this means two hidden layers with 4 nodes each
    for layer in layers:
        eachLayers["n_h" + str(i)] = layer
        i += 1 
    return eachLayers
```

```python
def initializeParameters(L, eachLayers):
    params = {}
    nprev = eachLayers["n_x"]
    
    for i in range(1, L):
        
        if(i < L-1):
            ncurrent = eachLayers["n_h" + str(i)]
        else: 
            ncurrent = eachLayers["n_y"]
            
        np.random.seed(42)
        params["W" + str(i)] = np.random.randn(ncurrent, nprev) * 0.01
        params["b" + str(i)] = np.zeros((ncurrent, 1))

        nprev = ncurrent
        
    return params
```

```python
def activationFunction(Z):
    expZ = np.exp(Z)
    sumExp = np.sum(expZ, axis = 0)
    
    return expZ/sumExp
```

```python
def sigmoid(Z):
    return 1/(1+np.exp(-Z))
```

```python
def linearForward(W, Aprev, b):
#     print(W.shape, Aprev.shape)
    Z = np.dot(W, Aprev) + b
    return Z
```

```python
def NNModelForward(L, X, params):
    Aprev = X
    finalValue = {
        "A0": X
    }
    
    for i in range(1, L-1):
        Z = linearForward(params["W" + str(i)], Aprev, params["b"+str(i)])
        A = sigmoid(Z)
        finalValue["Z"+str(i)] = Z
        finalValue["A"+ str(i)] = A
        Aprev = A
        
    Zl = linearForward(params["W" + str(L-1)], Aprev, params["b"+str(L-1)])
    Al = activationFunction(Zl)
    finalValue["ZL"]= Zl
    finalValue["AL"] = Al
    
    return finalValue    
```

```python
def computeCost(Y, Ycap):
    m = Y.shape[1]
#     print(Y.shape, Ycap.shape)
#     Y = Y[:, 0:5]
    losses = -1 * np.sum(Y * np.log(Ycap), axis = 0)
#     print(losses.shape)
    cost = np.sum(losses, axis = 0)/m
    return cost
```

```python
def gradientSigmoid(Z):
    return np.exp(-Z)/np.square(1+np.exp(-Z))
```

```python
def backwardPass(L, Y, linearForwardValues, params, m):
    grads = {}
    
    for i in range(L, 1, -1):
        
        if i == L:
            
            AL = linearForwardValues["AL"]
            grads["dZ" + str(i)] = (AL - Y)
            grads["dW" + str(i)] = 1/m * np.dot(grads["dZ" + str(i)], linearForwardValues["A"+str(L-2)].T)
            grads["db" + str(i)] = 1/m * np.sum(grads["dZ" +str(i)], axis = 1, keepdims = True)
#             print("Here2")
#             print(grads["dW"+str(i)].shape)
            
        else:
            
           
            grads["dA" + str(i)] = np.dot(params["W" + str(i)].T, grads["dZ" + str(i+1)])
            grads["dZ" + str(i)] = grads["dA" + str(i)] * gradientSigmoid(linearForwardValues["Z" + str(i-1)])
            grads["dW"+ str(i)] = 1/m * np.dot(grads["dZ" + str(i)], linearForwardValues["A" + str(i-2)].T)
            grads["db" + str(i)] = 1/m * np.sum(grads["dZ" + str(i)], axis = 1, keepdims = True)
        
#         print(" dZ: ", grads["dZ" + str(i)])
#         print(" dW: ", grads["dW" + str(i)])
#         print(" db: ", grads["db" + str(i)])   
        
    return grads
```

```python
def updateParams(L, grads, params, learningRate):
    for i in range(1, L):
#         print(params["W"+str(i)].shape)
#         print(grads["dW"+str(i+1)].shape)
#         print(params["b"+str(i)].shape)
#         print(grads["db"+str(i+1)].shape)
        
        params["W" + str(i)] = params["W" + str(i)] - learningRate * grads["dW" + str(i+1)]
        params["b" + str(i)] = params["b" + str(i)] - learningRate * grads["db" + str(i+1)]
        
    return params
```

```python
def NNModelBackward(linearForwardValues,Y, m, L, params, learningRate):
    grads = backwardPass(L, Y, linearForwardValues, params,m)
    return grads
```

```python
def Model(X,Y,num_iterations, learningRate):

#     print(pooledTrain.shape)

    eachLayers = initialization([4,4], X, trainY)
    L =     L = len(eachLayers)
    m = pooledTrain.shape[1]
    costs = []

    params = initializeParameters(L, eachLayers)

    for i in range(num_iterations+1):
        finalValue = NNModelForward(L, X, params)
        cost = computeCost(trainY, finalValue["AL"])
        grads = NNModelBackward(finalValue, Y, m, L, params, learningRate)
        params = updateParams(L, grads,params, learningRate)
        cost = np.squeeze(cost)
        costs.append(cost)
        if i % 1000 == 0:
            print ("Cost after %i iterations : %f" %(i, cost))
            
    return params
```

```python
trainedParams = Model(pooledTrain, trainY, 40000, 0.1)
```

```python
def predict(L, predictX, params):
    predictedValues = NNModelForward(L, predictX, params)
    return predictedValues
```

```python
def findAccuracy(actualValue, predictedValue, printText = "given"):
    predictedY = predictedValue["AL"]
    maxPositions= np.argmax(predictedY, axis = 0)
    comparision = (maxPositions == actualValue[0])
    count = np.count_nonzero(comparision)
    print("Accuracy of the predicted model for the %s data is %f" %(printText, (count/(predictedValue["AL"].shape[1])) * 100) + ' %')
```
